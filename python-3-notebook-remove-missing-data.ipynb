```python
#Importuoju savo rats duomenis
data_all = pd.read_csv("/home/emilija/Dokumentai/Magistras/SinBio/Sinbio-projektas/Žiurkių duomenys.csv")
```

```python
#nope
new_data = data_all.RatPyV1.dropna(inplace=True)
```

```python
#puiku, pasalina tik tas eiles, kuriose NA yra butent RatPyV1 arba RatPyV2 stulpeliuose 
new_data = data_all.dropna(subset=['RatPyV1', 'RatPyV2'])
```

```python
import fancyimpute
print("Fancyimpute imported successfully")
from fancyimpute import IterativeImputer
import numpy
```

```python
#data frame with the missings. I keep only floats
df_numeric = new_data.select_dtypes(include=[numpy.float])
```

```python
#imputuoja dataseto NA reiksmes IterativeImputer: A strategy for imputing missing values by modeling each feature 
#with missing values as a function of other features in a round-robin fashion. 
ip_data=IterativeImputer().fit_transform(df_numeric)
```

```python
ip_data
```

```python
#sutvarko IterativeImputer().fit_transform(df_numeric) outputa i normalia lentele su pavadinimais, kurie buvo pries tai
data=pd.DataFrame(ip_data, columns=df_numeric.columns, index=df_numeric.index)
```

```python
data
```

```python
#updeitina dataframe new_data reiksmemis is dataframe data. Reikia, kad sutaptu stulpeliu pavadinimai, 
#tada pakeicia new_data stulpelius su tais pavadinimais i data stulpelius su tais paciais pavadinimais
new_data.update(data)
```
